```python
import pretty_midi
import numpy as np
import mir_eval.display
import librosa.display
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = "/Volumes/Storage/goinfre/ademenet/midi_file/130000_Pop_Rock_Classical_Videogame_EDM_MIDI_Archive[6_19_15]/0/Celine Dion - Fly 2.mid"
```

```python
# Function from pretty_midi tutorial (http://nbviewer.jupyter.org/github/craffel/pretty-midi/blob/master/Tutorial.ipynb)
def plot_piano_roll(pm, start_pitch, end_pitch, fs=100):
    # Use librosa's specshow function for displaying the piano roll
    librosa.display.specshow(pm.get_piano_roll(fs)[start_pitch:end_pitch],
                             hop_length=1, sr=fs, x_axis='time', y_axis='cqt_note',
                             fmin=pretty_midi.note_number_to_hz(start_pitch))
```

```python
mid = pretty_midi.PrettyMIDI(filepath)
```

```python
plt.figure(figsize=(12, 4))
plot_piano_roll(mid, 24, 84)
```
